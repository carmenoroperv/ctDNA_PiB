```r
library(tidyverse)
```

```r
df = readRDS("../data/ATAC_predictions_train_20_predict_80_and_cases/ATAC_pred_lasso_std_TRUE_gaussian_trimmed_formatted_standardized.rds")

samples <- df %>% select(sample)
df <- df %>% select(-sample)
head(df)
```

```r
sample_types <- read.table("../data/sample_types.txt", header = F, sep = " ")
colnames(sample_types) <- c("sample", "sample_type")
head(sample_types)
```

```r
samples <- merge(samples, sample_types, by="sample")
head(samples)

samples <- samples %>% mutate(sample_type01 = ifelse(sample_type == "Healthy", "H", "C"))
head(samples)
```

```r
pca <- prcomp(~., df, scale = F)
```

```r
k = 2
km <- kmeans(df, centers=k, nstart=20)

plot <- ggplot(data = tibble(pc1 = pca$x[, 1], pc2 = pca$x[, 2], names = as.factor(km$cluster)), 
                    aes(pc1, pc2, color = samples$sample_type01)) + 
    geom_point() + 
    theme_minimal() +
    labs(title = paste("k = ", k, sep = ""), x = "PC1", y = "PC2", color = "Clusters")

print(plot)
```

```r
#dist(df) # 60 x 60 distance matrix (only lower triangle) is created

a <- plot(hclust(dist(df), method = "complete"), labels = samples$sample_type01, hang = 0, main = "Method: Complete")
print(a)
```

```r
b <- plot(hclust(dist(df), method = "single"), labels = samples$sample_type01, hang = 0, main = "Method: Single")
print(b)
```

```r
c <- plot(hclust(dist(df), method = "average"), labels = samples$sample_type01, hang = 0, main = "Method: Average")
print(c)
```

```r
d <- plot(hclust(dist(df), method = "centroid"), labels = samples$sample_type01, hang = 0, main = "Method: Centroid")
print(d)
```
