# Boosting on data smoothed with Gaussian (n = 10)

```r
library(tidyverse)

library(glmnet)

#install.packages("gbm")
library(gbm)
```

```r
sum_control_ATAC_bin_rm_gaussian_mean <- read.csv("data/sum_control_normalized_gaussian_smooth_ATAC_bin_rm.csv")

sum_control_ATAC_bin_rm = read.csv("data/sum_control_normalized_ATAC_bin_rm.csv")
ATAC_value <- pull(sum_control_ATAC_bin_rm, ATAC_val)

sum_control_ATAC_bin_rm_gaussian_mean$ATAC_val = ATAC_value
head(sum_control_ATAC_bin_rm_gaussian_mean)
```

```r
set.seed(0)

fit <- gbm(formula  = ATAC_val ~ ., 
               data = sum_control_ATAC_bin_rm_gaussian_mean, 
            n.trees =500, 
           cv.folds = 10)
```

```r
fit
```

## Plot the training error and the CV RMSE

```r
pd <- tibble(rmse_cv = sqrt(fit$cv.error), 
             rmse_train = sqrt(fit$train.error)) %>%
  mutate(tree = row_number()) %>%
  pivot_longer(names_to = "key", values_to = "value", -tree)

ggplot(pd, aes(x=tree, y=value, color=key)) + 
  geom_line() + 
  geom_point() + 
  NULL
```

```r
gbm.perf(fit, method = "cv")
```

```r
fit$train.error[which.min(sqrt(fit$train.error))]

fit$cv.error[which.min(sqrt(fit$cv.error))]
```

```r
#predictions <- as.data.frame(predict(fit, sum_control_ATAC_bin_rm))

set.seed(0)

fit <- gbm(formula  = ATAC_val ~ ., 
               data = sum_control_ATAC_bin_rm_gaussian_mean, 
            n.trees =500, 
           cv.folds = 10)


observed <- as.data.frame(sum_control_ATAC_bin_rm$ATAC_val)
```

```r
predicted <- fit$cv.fitted
combined <- cbind(predicted, observed)
colnames(combined) <- c("predicted", "observed")
head(combined)
```

```r
ggplot(data = combined, aes(x = observed, y = predicted)) + 
    geom_point(size = 0.5) + 
    geom_smooth(method = "lm", formula = y~x)
```

```r
cor(combined$observed, combined$predicted)
```
